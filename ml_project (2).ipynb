Filling the values with the mean

```python
data['ALB'].fillna(data['ALB'].mean(), inplace=True)
data['ALP'].fillna(data['ALP'].mean(), inplace=True)
data['CHOL'].fillna(data['CHOL'].mean(), inplace=True)
data['PROT'].fillna(data['PROT'].mean(), inplace=True)
data['ALT'].fillna(data['ALT'].mean(), inplace=True)
print(data.isnull().sum())
```

Converting categorical values into numerical values using OneHotEncoder


```python
from sklearn.preprocessing import OneHotEncoder, LabelEncoder,StandardScaler
df = OneHotEncoder()
df=['Sex']
data=pd.get_dummies(data,columns=df)
```

```python
data['Category'] = data['Category'].replace({'0=Blood Donor': 0, '0s=suspect Blood Donor': 0, '1=Hepatitis': 1, '2=Fibrosis': 1, '3=Cirrhosis': 1})
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
y_train
```

Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier, export_text
cf = DecisionTreeClassifier(criterion='entropy')
cf.fit(x,y)
tr = export_text(cf,feature_names=list(x.columns))
print(tr)
```

```python
#generate and print the textual represenatation of the decision tree
import matplotlib.pyplot as plt
from sklearn import tree
fig, ax= plt.subplots(figsize=(10,10))
tree.plot_tree(cf,fontsize=10)
plt.show()
```

```python
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
y1_pred=classifier.predict(x_test)
print("Accuracy Score : ",accuracy_score(y_test,y1_pred)*100)
```

```python
cnf_matrix = confusion_matrix(y_test,y1_pred)
plt.figure(figsize=(4,3))
p = sns.heatmap((cnf_matrix),annot=True,cmap="Blues",fmt='g')
plt.title("Confusion Matrix ",y=1.1)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
print(classification_report(y_test,y1_pred))
```

The .fit() method in scikit-learn is used to train a machine learning model on a given dataset. It takes the input features (X) and corresponding target values (y) as arguments and fits the model to the training data.

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
#n_estimators: Number of trees in the forest.
#max_depth: Maximum depth of the tree.
#min_samples_split: Minimum number of samples required to split an internal node
param_grid = { 'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], }
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=10)
grid_search_rf.fit(x_train, y_train)
```

```python
best_params_for_rf = grid_search_rf.best_params_
print(best_params_for_rf)
```

```python
# Fitting Random Forest with the best parameters
model = RandomForestClassifier(**best_params_for_rf, random_state=42)
model.fit(x_train, y_train)
```

```python
y3_pred = model.predict(x_test)
print("Accuracy Score  : ",accuracy_score(y_test, y3_pred) * 100)
```

```python
cnf_matrix = confusion_matrix(y_test,y3_pred)
plt.figure(figsize=(4,3))
p = sns.heatmap((cnf_matrix),annot=True,cmap="Blues",fmt='g')
plt.title("Confusion Matrix of RandomForest",y=1.1)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
print(classification_report(y_test,y3_pred))
```

KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
test_scores=[]
train_scores=[]
for i in range(1,15):
  knn=KNeighborsClassifier(i)
  knn.fit(x_train,y_train)
  train_scores.append(knn.score(x_train,y_train))
  test_scores.append(knn.score(x_test,y_test))
```

```python
train_scores
```

```python
test_scores
```

```python
max_train_score = max(train_scores)
train_scores_ind = [i for i,v in enumerate(train_scores) if v==max_train_score]
print("Max train score = ",max_train_score*100," and k = ",list(map(lambda x:x+1,train_scores_ind)))
```

```python
max_test_score = max(test_scores)
test_scores_ind = [i for i,v in enumerate(test_scores) if v==max_test_score]
print("Max test score = ",max_test_score*100," and k = ",list(map(lambda x:x+1,test_scores_ind)))
```

```python
plt.figure(figsize=(12,5))
p = sns.lineplot(x=range(1,15), y=train_scores,marker='*',label='Train Score')
p = sns.lineplot(x=range(1,15), y=test_scores,marker='o',label='Test Score')
plt.show()
```

```python
knn=KNeighborsClassifier(12)
knn.fit(x_train,y_train)
knn.score(x_test,y_test)
```

```python
y5_pred = knn.predict(x_test)
cnf_matrix = confusion_matrix(y_test,y5_pred)
plt.figure(figsize=(4,3))
p = sns.heatmap((cnf_matrix),annot=True,cmap="Blues",fmt='g')
plt.title("Confusion Matrix of KNN",y=1.1)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```

```python
print("Accuracy Score of KNN : ",accuracy_score(y_test,y5_pred)*100)
print(classification_report(y_test,y5_pred))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
from sklearn.model_selection import GridSearchCV
#n_estimators: Number of trees in the forest.
#max_depth: Maximum depth of the tree.
#min_samples_split: Minimum number of samples required to split an internal node
param_grid = { 'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10], }
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=10)
grid_search_rf.fit(x_train, y_train)
best_params_for_rf = grid_search_rf.best_params_
print(best_params_for_rf)
# Fitting Random Forest with the best parameters
model = RandomForestClassifier(**best_params_for_rf, random_state=42)
model.fit(x_train, y_train)
y_predict = model.predict(x_test)
print("Accuracy Score  : ",accuracy_score(y_test, y_predict) * 100)
cnf_matrix = confusion_matrix(y_test,y_predict)
plt.figure(figsize=(4,3))
p = sns.heatmap((cnf_matrix),annot=True,cmap="Blues",fmt='g')
plt.title("Confusion Matrix of RandomForest",y=1.1)
plt.ylabel('Actual Label')
plt.xlabel('Predicted Label')
```
